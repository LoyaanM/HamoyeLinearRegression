```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('energydata_complete.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
simple_linear_reg_df = df[['T2', 'T6']].sample(20, random_state=2)
sns.regplot(x='T2', y='T6', data=simple_linear_reg_df)
```

```python
df2=df.drop(['date', 'lights'], axis=1)
df2
```

```python
df2.describe()
df2.isnull().sum()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
#Normalising dataset to a common scale using min max scaler
scaled_df2 = pd.DataFrame(scaler.fit_transform(df2), columns=df2.columns)
features_df2 = scaled_df2.drop(columns=['Appliances'])
appliances_target = scaled_df2['Appliances']
```

```python
#Splitting into Training and Testing dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(features_df2, appliances_target, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model_linear = LinearRegression()
#fit model to the training dataset
model_linear.fit(x_train, y_train)
```

```python
#obtain predictions
predicted_values = model_linear.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, predicted_values)
print('mae is:', round(mae, 2))
```

```python
import numpy as np
rss = np.sum(np.square(y_test - predicted_values))
print('rss is:', round(rss, 2))
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
print('rmse is:', round(rmse, 3))
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y_test, predicted_values)
print('coefficient of determination is:', round(r2, 2))
```

```python
def get_weights_df2(model, feat, col_name): #returns weight of every feature
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df2 = pd.DataFrame(weights).reset_index()
    weights_df2.columns = ['Features', col_name]
    weights_df2[col_name].round(3)
    return weights_df2
```

```python
linear_model_weights = get_weights_df2(model_linear, x_train, 'LinearModel_Weight')
linear_model_weights
```

```python
#Ridge Regression
from sklearn.linear_model import Ridge
ridge_reg = Ridge(alpha=0.4)
ridge_reg.fit(x_train, y_train)
```

```python
pred_ridge = ridge_reg.predict(x_test)
rmse = np.sqrt(mean_squared_error(y_test, pred_ridge))
print('rmse ridge is:', round(rmse, 3))
```

```python
from sklearn.linear_model import Lasso
lasso_reg = Lasso(alpha=0.001)
lasso_reg.fit(x_train, y_train)
```

```python
def get_weights_df2(model, feat, col_name): #returns weight of every feature
    weights = pd.Series(model.coef_, feat.columns).sort_values()
    weights_df2 = pd.DataFrame(weights).reset_index()
    weights_df2.columns = ['Features', col_name]
    weights_df2[col_name].round(3)
    return weights_df2
```

```python
lasso_weights_df2 = get_weights_df2(lasso_reg, x_train, 'Lasso_Weight' )
lasso_weights_df2
```

```python
pred_lasso = lasso_reg.predict(x_test)
rmse = np.sqrt(mean_squared_error(y_test, pred_lasso))
print('rmse lasso is:', round(rmse, 3))
```
